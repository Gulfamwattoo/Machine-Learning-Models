```python
X=df.drop(["Class"], axis=1)
y=df["Class"]
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, shuffle=True)
```

```python
# Standard Scaler
scaler= StandardScaler()
X_train_scaler=scaler.fit_transform(X_train)
X_test_scaler=scaler.transform(X_test)
```

## ANN


```python
# create tensor
import torch 
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
```

```python
X_train_tensor=torch.tensor(X_train_scaler, dtype=torch.float32)
y_train_tensor=torch.tensor(y_train, dtype=torch.long)

X_test_tensor=torch.tensor(X_test_scaler, dtype=torch.float32)
y_test_tensor=torch.tensor(y_test, dtype=torch.long)
```

```python
#Tensor Dataset
train_dataset=TensorDataset(X_train_tensor,y_train_tensor)
test_dataset= TensorDataset(X_test_tensor,y_test_tensor)
```

```python
train_load= DataLoader(train_dataset, batch_size=32, shuffle=True )
test_load= DataLoader(test_dataset , batch_size=32)
```

```python
# Buid model
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()

        self.model=nn.Sequential(
            nn.Linear(X.shape[1],64),
            nn.ReLU(),

            nn.Linear(64,64),
            nn.ReLU(),

            nn.Linear(64,7)
            
        
        
        )

    def forward(self,x):
        return self.model(x)
        

        
```

```python
model=ANN()
# loss and optimiser
criteria=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters())
```

```python
# Training Model 

epochs=50
for epoch in range(epochs):
    model.train()
    runing_loss=0.0
    for xb, yb in train_load:
        optimizer.zero_grad()
        outputs=model(xb)
        
        loss= criteria(outputs,yb)
        loss.backward()
        optimizer.step()

        runing_loss+=loss.item()
    training_loss=runing_loss/len(train_load)

    print(f"Epochs ={epoch+1}/{epochs} and loss is {training_loss}")
    
```

```python
#Evaluate
model.eval()
total=0
correct=0
with torch.no_grad():
    for xb, yb in test_load:
        outputs=model(xb)
        _,predicted=torch.max(outputs,1)
        correct+=(predicted==yb).sum().item()
        total+=yb.size(0)

    print("Accuracy:" , correct/total*100)
        
        
```
